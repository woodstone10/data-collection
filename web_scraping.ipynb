# Web Scraping

---
extracting contents from web pageb (https://basicenglishspeaking.com/daily-english-conversation-topic)

easily click on links for all contents (ex, http://basicenglishspeaking.com/wp-content/uploads/audio/QA/Q-01-01.mp3)




```python
import requests
from bs4 import BeautifulSoup
```

```python
def crawl(url):
    try:
        rData = requests.get(url)  
        print(rData)  
        return rData
    except:
        print("<Response Error>")
        return 0

def parse(str):
    bsData = BeautifulSoup(str, "html.parser")  
    links = []
    for link in bsData.findAll('a'):
        links.append(link.get('href')) #get url link
    return links
```

```python
str = crawl('https://basicenglishspeaking.com/daily-english-conversation-topic')
url = parse(str.text)
url
```

## Web Scraping contents

```python
def contents(str):
    try:
        bsData = BeautifulSoup(str.text, "html.parser")
        contents = bsData.findAll("div", {"class": "sc_player_container1"})
        for i in range(len(contents)):
            print(contents[i])
    except:
        print("No contents")

for i in range(len(url)):
    print(url[i])
    str = crawl(url[i])
    contents(str)
```
